```python
def pearson(s1, s2):
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))
```

```python
def get_recs(movie_name, M, num):

    import numpy as np
    reviews = []
    for title in M.columns:
        if title == movie_name:
            continue
        cor = pearson(M[movie_name], M[title])
        if np.isnan(cor):
            continue
        else:
            reviews.append((title, cor))
    
    reviews.sort(key=lambda tup: tup[1], reverse=True)
    return reviews[:num]
```

```python
print ("Pearson Correlation coefficient of :")
a= []
for i in range(0, 16, 1):
    cor=pearson(M.ix[120], M.ix[i])
    print ("(" + M.index[120] + " , " + M.index[i] + ") :", cor)
    a.append(cor)
```

```python
def pred_rate(x, a):
    j=0
    k=0
    for i in range(0, 16, 1):
        if x.irow(i)>0 and a[i]>0: 
            j= j+(a[i]*x.irow(i))
            k= k+a[i]
        else:
            continue
    return j/k
```

```python
print ("For all the movies I haven't seen, using the class reviews, my ratings are:")
for i in range(0, 4, 1):
    x=M[not_rated.index[i]]
    rate=pred_rate(x, a)
    print (not_rated.index[i] +": ", rate.round(1))
```
