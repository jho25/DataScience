```python
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.datasets import fetch_20newsgroups
categories = ['rec.sport.baseball']
dataset = fetch_20newsgroups(subset='all',shuffle=True, random_state=42, categories=categories)
corpus = dataset.data
```

```python
stopset = set(stopwords.words('english'))
stopset.update(['lt','p','/p','br','amp','quot','field','font','normal','span','0px','rgb','style','51', 
                'spacing','text','helvetica','size','family', 'space', 'arial', 'height', 'indent', 'letter'
                'line','none','sans','serif','transform','line','variant','weight','times', 'new','strong', 'video', 'title'
                'white','word','letter', 'roman','0pt','16','color','12','14','21', 'neue', 'apple', 'class', '\n', '@', '00', '00 00',
               'nntp', 'edu', 'cs','com','000','etc', 'john', 'morris', 'mjones', 'jones','kingman', 'jack','dave','williams', 'ted', 'duke', 'hirschbeck', 'asd'
               ,'01','02','03','04','05', '06', '07', '08', '09', '10','vb30','samuel','torre', 'scott', 'barman','smith','steve'
                ,'gant','hall','david', 'sabo', 'aix','clemens','lopez','roger','viola','mike', 'joe', 'jays','braves','fischer','sherri','sreph','mark','nichols',
               'dale','stehpenson','michael','lowenstein' 'barry', 'aaa','alomar','boggs','pitt','greg','bob','000th','001','0010','edward'])
```

```python
vectorizer = TfidfVectorizer(stop_words=stopset,
                                 use_idf=True, ngram_range=(1, 3))
X = vectorizer.fit_transform(corpus)
```

```python
X[0]
```

```python
print (X[0])
```

```python
X.shape
```

```python
lsa = TruncatedSVD(n_components=150, n_iter=100)
lsa.fit(X)
```

```python
lsa.components_[0]
```

```python
import sys
print (sys.version)
```

```python
terms = vectorizer.get_feature_names()
for i, comp in enumerate(lsa.components_): 
    termsInComp = zip (terms,comp)
    sortedTerms =  sorted(termsInComp, key=lambda x: x[1], reverse=True) [:10]
    print("Concept %d:" % i )
    for term in sortedTerms:
        print(term[0])
    print (" ")
```
